```python
# Task 2: Multi-line plot of average MPG over years by origin
avg_mpg_by_year_origin = df.groupby(['model_year', 'origin'])['mpg'].mean().unstack()
plt.figure(figsize=(10, 6))
for origin in ['japan', 'europe', 'usa']:
    plt.plot(avg_mpg_by_year_origin.index, avg_mpg_by_year_origin[origin], label=origin.capitalize(), marker='o')
plt.title('Evolution of Average MPG by Origin (1970-1982)')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.show()
print("Task 2: Multi-line plot displayed")
print("Explanation: The plot shows average MPG trends over model years for Japan, Europe, and USA, highlighting fuel efficiency improvements.")
print()
```

```python
# Task 3: Seaborn scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df.reset_index(), x='horsepower', y='weight', hue='origin', size='mpg',
                hue_order=['japan', 'europe', 'usa'], sizes=(20, 200))
plt.title('Horsepower vs Weight by Origin, Sized by MPG')
plt.xlabel('Horsepower')
plt.ylabel('Weight (lbs)')
plt.legend(title='Origin')
plt.show()
print("Task 3: Scatterplot displayed")
print("Explanation: The scatterplot visualizes horsepower vs weight, colored by origin and sized by MPG, with Japan, Europe, USA in specified hue order.")
print()
```

```python
 #Task 4: Cars with multiple appearances and std(mpg) < 1.0
# Group by name and count unique model years
car_years = df.groupby('name')['model_year'].nunique()
multi_year_cars = car_years[car_years > 1].index
# Compute std and mean mpg for these cars
mpg_stats = df[df['name'].isin(multi_year_cars)].groupby('name').agg(
    appearances=('model_year', 'nunique'),
    mean_mpg=('mpg', 'mean'),
    std_mpg=('mpg', 'std')
)
# Filter for std(mpg) < 1.0
low_var_cars = mpg_stats[mpg_stats['std_mpg'] < 1.0][['appearances', 'mean_mpg']].reset_index()
# Sort by appearances (descending) and mean_mpg (descending)
low_var_cars = low_var_cars.sort_values(by=['appearances', 'mean_mpg'], ascending=[False, False])
print("Task 4: Cars with multiple appearances and std(mpg) < 1.0")
if not low_var_cars.empty:
    print("Model Name | Appearances | Average MPG")
    print("-" * 40)
    for _, row in low_var_cars.iterrows():
        print(f"{row['name']} | {row['appearances']} | {row['mean_mpg']:.2f}")
else:
    print("No cars with multiple appearances have std(mpg) < 1.0")
print("Explanation: Identified cars appearing in multiple model years, computed their MPG standard deviation, and filtered for std(mpg) < 1.0, sorted by appearances and MPG.")
```
