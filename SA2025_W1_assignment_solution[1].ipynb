### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
plt.figure(figsize=(12, 6)) #size

for origin in df['origin'].unique():
    origin_data = df[df['origin'] == origin]
    avg_mpg = origin_data.groupby('model_year')['mpg'].mean()
    plt.plot(avg_mpg.index + 1900, avg_mpg.values, marker='o', label = origin.capitalize())

#plotting

plt.title('Evolution of <MPG> Over the Years by Origin')
plt.xlabel('Year')
plt.ylabel('<MPG>')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
df_original = pd.read_csv('Cars.csv')

# cars with multiple years

car_year_counts = df_original.groupby('name')['model_year'].nunique()
multi_year_cars = car_year_counts[car_year_counts > 1].index

# mpg -> standard deviation 
consistent_cars = []
for car in multi_year_cars:
    car_data = df_original[df_original['name'] == car]
    mpg_std = car_data['mpg'].std()
    if mpg_std < 1.0:
        consistent_cars.append({
            'name': car,
            'appearances': len(car_data),
            'mpg_std': mpg_std,
            'avg_mpg': car_data['mpg'].mean()
        })


consistent_df = pd.DataFrame(consistent_cars)
if not consistent_df.empty:
    consistent_df = consistent_df.sort_values(['appearances', 'avg_mpg'], ascending=[False, False])
    print("Most consistent car models:")
    print(consistent_df)
else:
    print("No cars found with std(mpg) < 1.0 across multiple years")
```
